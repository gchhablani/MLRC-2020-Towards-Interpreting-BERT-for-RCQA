```python
squad_train = '../data/SQuAD-explorer/dataset/train-v1.1.json'
squad_dev = '../data/SQuAD-explorer/dataset/dev-v1.1.json'
```

```python
with open(squad_dev,'r') as f:
    dev_s = json.load(f)
```

```python
with open(squad_train,'r') as f:
    train_s = json.load(f)
```

```python
squad_train
```

```python
train_s['data'][0]
```

```python
word_lens = []
token_lens = []
count = 0
for sample in train_s['data']:
    for paragraph in sample['paragraphs']:
        word_lens.append(len(paragraph['context'].split()))
        token_lens.append(len(bt.tokenize(paragraph['context'])))
        count+=len(paragraph['qas'])
```

```python
np.mean(word_lens),np.max(word_lens), np.mean(token_lens), np.max(token_lens), count
```

```python
word_lens = []
token_lens = []
count = 0
for sample in dev_s['data']:
    for paragraph in sample['paragraphs']:
        word_lens.append(len(paragraph['context'].split()))
        token_lens.append(len(bt.tokenize(paragraph['context'])))
        count+=len(paragraph['qas'])
```

```python
np.mean(word_lens),np.max(word_lens), np.mean(token_lens), np.max(token_lens), count
```

```python
count = 0
count_0 = 0
count_2 = 0
common = 0
for sample in train_s['data']:
    for paragraph in sample['paragraphs']:
        for qa in paragraph['qas']:
            if (len(qa['answers'])==0):
                count_0+=1
            if len(qa['answers'])>1:
                count_2+=1
                print(qa)
            if len(qa['answers'])>0 and paragraph['context'].find(qa['answers'][0]['text'])==-1:
                count+=1
                
            if len(qa['answers'])>1 and len(qa['answers'])>0 and paragraph['context'].find(qa['answers'][0]['text'])==-1:
                count+=1
                
print(count,count_2, common, count_0)
```

```python
count = 0
count_0 = 0
count_2 = 0
common = 0
for sample in dev_s['data']:
    for paragraph in sample['paragraphs']:
        for qa in paragraph['qas']:
            if (len(qa['answers'])==0):
                count_0+=1
            if len(qa['answers'])>1:
                flag = False
                answer = qa['answers'][0]
                for i in qa['answers'][1:]:
                    if(answer!=i):
                        flag = True
                count_2+=int(flag)
                if(flag):
                    print(qa)
            if len(qa['answers'])>0 and paragraph['context'].find(qa['answers'][0]['text'])==-1:
                count+=1
                
            if len(qa['answers'])>1 and len(qa['answers'])>0 and paragraph['context'].find(qa['answers'][0]['text'])==-1:
                count+=1
                
print(count,count_2, common, count_0)
```
